# Crear una regresion lineal con estatura como variable explicatoria y peso como variable explicada

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Load the data
url = 'http://wiki.stat.ucla.edu/socr/index.php?title=SOCR_Data_Dinov_020108_HeightsWeights&oldid=12529'
height_weight_df = pd.read_html(url)[1][['Height(Inches)','Weight(Pounds)']]

# Drop any missing values
height_weight_df = height_weight_df.dropna()

# Define the explanatory and response variables
X = height_weight_df[['Height(Inches)']]
y = height_weight_df['Weight(Pounds)']

# Fit the regression model
model = LinearRegression()
model.fit(X, y)

# Get the model parameters
intercept = model.intercept_
slope = model.coef_[0]

# Make predictions
y_pred = model.predict(X)

```

# 3. Hallar el intercepto

```python
# Obtener los par√°metros del modelo
intercept = model.intercept_
slope = model.coef_[0]

# Hacer predicciones
y_pred = model.predict(X)

# Mostrar el intercepto y la pendiente
print(f'Intercepto: {intercept}')
print(f'Pendiente: {slope}')
```

# 4. Hacer la grafica

```python
# Plot the regression line
plt.scatter(height_weight_df['Height(Inches)'], height_weight_df['Weight(Pounds)'], label='Data points')
plt.plot(height_weight_df['Height(Inches)'], y_pred, color='red', label='Regression line')
plt.xlabel('Height (Inches)')
plt.ylabel('Weight (Pounds)')
plt.legend()
plt.show()
```

# 5. Hayar la correlacion y valor P


```python
correlation, p_value = pearsonr(height_weight_df['Height(Inches)'], height_weight_df['Weight(Pounds)'])

float(correlation), float(p_value)
```
